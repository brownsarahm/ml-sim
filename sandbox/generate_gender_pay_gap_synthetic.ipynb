```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
summarized_gpg_df = pd.read_csv('gender_pay_gap.csv')
summarized_gpg_df
```

```python
def normal_expand_df(row,count_col, sample_col,sigma):
    """
    expand a row of a summarized table
    """
    retain_cols = [c for c in row.index if c not in [sample_col,count_col]]
    n_repeats = row[count_col]
    row_df = row[retain_cols].to_frame().T
    repeated_df = pd.concat([row_df]*row[count_col], ignore_index=True,axis=0)
    repeated_df[sample_col] = np.random.normal(row[sample_col],sigma,
                                               row[count_col])

    return repeated_df
    
```

```python
expand_headcount_5noise = lambda row: normal_expand_df(row,'Headcount','MeanHourlyPay',5)
a = summarized_gpg_df.apply(expand_headcount_5noise,axis=1)
expanded_gpg_df = pd.concat([ai for ai in a])
expanded_gpg_df.rename(columns={'MeanHourlyPay':'hourlyPay'},inplace=True)
expanded_gpg_df.head()
```
